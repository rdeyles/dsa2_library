```python
import matplotlib.pyplot as mp
import networkx as nx
```

```python
G=nx.Graph()
nodes=['Paris','Nantes','Lyon','Marseille']
G.add_nodes_from(nodes)
G.add_edge('Paris','Nantes',weight=34)
G.add_edge('Paris','Marseille',weight=45)
G.add_edge('Paris','Lyon',weight=33)
G.add_edge('Nantes','Lyon',weight=29)
G.add_edge('Nantes','Marseille',weight=41)
G.add_edge('Marseille','Lyon',weight=15)
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos,style='dotted')
nx.draw_networkx_labels(G,pos)
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
mp.show()
```
