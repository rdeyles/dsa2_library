```python
import matplotlib.pyplot as mp
import networkx as nx
from collections import deque
```

```python
G=nx.random_labeled_rooted_tree(10)
nx.draw(G,with_labels=True)
mp.show()
```

```python
source=1
seen = []
seen.append(source)
q=deque()
q.append(source)

while q:
    node=q.popleft()
    print(node)
    for n in G[node]:
        if n not in seen:
            seen.append(n)
            q.append(n)
```
